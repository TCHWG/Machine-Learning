```python
import os
import zipfile

def remove_extension(file):
    """Remove the file extension from a filename."""
    return ".".join(os.path.basename(file).split('.')[:-1])

def transcribe(file, outfolder='.'):
    """Transcribe the file using Transkun."""
    print(f"Processing: {file}")

    # Output path for the transcribed file
    output_path = os.path.join(outfolder, remove_extension(file) + ".mid")
    
    # Run the Transkun transcription command
    os.system(f'transkun "{file}" "{output_path}" --device cpu')
    
    return output_path

input_folder = "music"
output_folder = "result"

# Ensure output folder exists
os.makedirs(output_folder, exist_ok=True)

input_files = [os.path.join(input_folder, f) for f in os.listdir(input_folder)]

if len(input_files) >= 2:
    with zipfile.ZipFile(os.path.join(output_folder, "transcribed.zip"), 'w') as new_zip:
        for file_path in input_files:
            if os.path.isfile(file_path): 
                transcribed_file = transcribe(file_path, output_folder)
                new_zip.write(transcribed_file, arcname=os.path.basename(transcribed_file), compress_type=zipfile.ZIP_DEFLATED)
    print(f"Transcribed files saved as: {os.path.join(output_folder, 'transcribed.zip')}")
else:
    for file_path in input_files:
        if os.path.isfile(file_path):  
            transcribed_file = transcribe(file_path, output_folder)
            print(f"Transcribed file saved as: {transcribed_file}")
```
